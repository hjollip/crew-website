```python
import yfinance

# Create a Ticker object for the desired stock (e.g., Apple)
aapl = yf.Ticker("AAPL")
```

```python
# Get a list of all available options expiration dates
expirations = aapl.options
print(expirations)
```

```python
# Select the first expiration date
date = aapl.options[0]

# Get the option chain
opt = aapl.option_chain(date)

# Access calls and puts
calls = opt.calls
puts = opt.puts

# Display call options
print(calls)
```
